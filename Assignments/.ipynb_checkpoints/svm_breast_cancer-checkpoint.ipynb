```python
# ANALYSIS
import pandas as pd
import numpy as np

# VISUALISATION
import matplotlib.pyplot as plt
import seaborn as sns
# Allows charts to appear in notebook
%matplotlib inline

# MACHINE LEARNING
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics
```

# Section 1: *Data Analysis*

```python
cancer = load_breast_cancer()
print(cancer.DESCR)
target = cancer.target
data = cancer.data
features = cancer.feature_names

cancer_df = pd.DataFrame(data, columns=features,)
cancer_df["Infected"] = target
```

## SECTION 3: *Split Data & Scaling*

##  Data split


```python
X = np.array(data)
y = np.array(target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Data scaling

```python
standard_scaler = preprocessing.StandardScaler()
X_standard_scaler = standard_scaler.fit_transform(X_train)
X_test_standard_scaler = standard_scaler.fit_transform(X_test)

X_standard_scaler
```

## SECTION 4: *Support Vector Machine Model*

```python
svc = SVC(gamma='auto')
```

## Fit model

```python
svc.fit(X_standard_scaler, y_train)
```

## Prediction

```python
y_pred = svc.predict(X_test_standard_scaler)
y_pred
```

## Confusion Matrix

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)

confusion_matrix
```

### Calculate Accuracy of Confusion Matrix

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred)
f1 = metrics.f1_score(y_test, y_pred)
accuracy, precision, recall, f1
```
